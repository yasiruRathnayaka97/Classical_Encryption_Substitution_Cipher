# **Classical Encryption/Substitution Cipher**

**Import necessary libries to the code**

```
import matplotlib.pyplot as plt
import random
from bs4 import BeautifulSoup
from urllib.request import urlopen
from collections import OrderedDict
```

**create random cipher**

```
def create_cipher():
    ASCII_number_A=ord('A')
    ASCII_number_Z=ord('Z')
    number_of_english_letters=26
    random_no_duplicate_number_list=random.sample(range(ASCII_number_A,ASCII_number_Z+1),number_of_english_letters )
    cipher={}
    n=ASCII_number_A
    for num in random_no_duplicate_number_list:
      cipher[chr(n)]=chr(num)
      n+=1
    return cipher
    

```

```
cipher=create_cipher()
print (cipher)
```

**create a function to encrypt text**

```
def encrypt(text,cipher):
  encrypted_text=""
  for character in text:
    if(character.isalpha()):
       if(character.upper() in cipher):
          encrypted_text+=cipher[character.upper()]
       else:
           encrypted_text+=character
    else:
      encrypted_text+=character
  return encrypted_text
```

**function to extract text from URL**

```
def  extract_text_from_url(url):
  html = urlopen(url).read()
  soup = BeautifulSoup(html)
  for script in soup(["script", "style"]):
      script.decompose()
  strips = list(soup.stripped_strings)
  return " ".join(strips)
```

**function to find frequency of characters in encrypted text**

```
def frequency(text,cipher):
  frequency_dict={}
  for character in text:
    if (character in cipher):
      if(character in frequency_dict):
          frequency_dict[character]+=1
      else:
          frequency_dict[character]=0
  
  return dict(OrderedDict(sorted(frequency_dict.items(), key = lambda t: t[0])))       
  
```

```
frequency_dict=frequency(encrypted_wiki,cipher)
print (frequency_dict)
```

```
def max_n_frequency(d,n):
    return dict(OrderedDict(sorted(d.items(), key = lambda t: t[1],reverse=True)[:n]))
   
```

```

def match_original_characters_for_frequencies(frequencies,cipher):
  original_characters={}
  list_cipher_keys=list(cipher.keys())
  list_cipher_values=list(cipher.values())
  for character in frequencies:
       original_characters[character]=list_cipher_keys[list_cipher_values.index(character)]
  return  original_characters
```

**function to partial decryption**

```
def partial_decryption(encrypted_text,decrypt_characters,cipher):
  list_cipher_keys=list(cipher.keys())
  list_cipher_values=list(cipher.values())
  partial_decrypted_text=""
  for character in encrypted_text:
    if (character in decrypt_characters):
        partial_decrypted_text+=list_cipher_keys[list_cipher_values.index(character)]
    else:
         partial_decrypted_text+=character
  return partial_decrypted_text

```
